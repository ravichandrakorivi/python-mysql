## Employees hired in the last 75 days

```python
query = """
select *, datediff(curdate(), hire_date) as diff
from employeesTbl_hired_in_last_duration
order by hire_date desc
"""

cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

```python
query = """
select *, datediff(curdate(), hire_date) as diff
from employeesTbl_hired_in_last_duration
where datediff(curdate(), hire_date) <= 75
order by hire_date desc
"""

cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

```python
query = """
select *, timestampdiff(day, hire_date, curdate()) as diff
from employeesTbl_hired_in_last_duration
order by hire_date desc
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

```python
query = """
select *, timestampdiff(day, hire_date, curdate()) as diff
from employeesTbl_hired_in_last_duration
where timestampdiff(day, hire_date, curdate()) <= 75
order by hire_date desc
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

## Employees hired in the last 3 months

```python
query = """
select *
from employeesTbl_hired_in_last_duration
where hire_date >= '2026-01-12'
order by hire_date desc
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

```python
query = """
select curdate() - interval 3 month
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

```python
query = """
select *
from employeesTbl_hired_in_last_duration
where hire_date >= curdate() - interval 3 month
order by hire_date desc
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```

## Employees hired in the last 1 year

```python
query = """
select *,timestampdiff(year, hire_date, curdate()) as diff
from employeesTbl_hired_in_last_duration
where hire_date >= curdate() - interval 1 year
order by hire_date desc
"""
cursor.execute(query)
employees = cursor.fetchall()
headers = [tup[0] for tup in cursor.description]
display(pd.DataFrame(employees, columns=headers))
```
