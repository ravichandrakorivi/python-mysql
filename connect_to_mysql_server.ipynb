# Connect to MYSQL Server using `mysql.connector`

## Connect to MySQL Server

```python
import os
from dotenv import load_dotenv
import mysql.connector

load_dotenv()

conn = mysql.connector.connect(
    host=os.getenv("MYSQL_HOST"),
    user=os.getenv("MYSQL_USER"),
    password=os.getenv("MYSQL_PASSWORD"),
    port=os.getenv("MYSQL_PORT")
)

print("Connected to the MySQL server successfully!")
```

## Create a cursor (this is how you run SQL)

```python
cursor = conn.cursor()
```

## Run queries

```python
cursor.execute("SHOW DATABASES")
results = cursor.fetchall()
```

```python
for result in results[1:3]:
    print(result[0])
```

## `Cursor`

- The `mysql.connector.connect.cursor()` is an iterator.
- Once it’s fully consumed, it becomes empty.
- So on the second run, nothing prints as the `cursor` is already exhausted.

## Store `cursor` results

```python
cursor.execute("SHOW DATABASES")
results = cursor.fetchall()
```

```python
dbs = [result[0] for result in results]
dbs[1:3]
```
